```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
vs = [200,200,200]

ps_unity=[123,58,20]
ps_lag=[56,26,8]
ps_lead=[98,40,12]

vr_unity=[173,188,196]
vr_lag=[171,187,196]
vr_lead=[204,203,202]

pr_unity=[73,37,13]
pr_lag=[43,21,7]
pr_lead=[73,31,10]
```

```python
E_unity = np.divide(pr_unity, ps_unity)
Vreg_unity= np.divide(np.subtract(vs,vr_unity), vr_unity)
losses_unity = np.subtract(ps_unity,pr_unity)

E_lag = np.divide(pr_lag, ps_lag)
Vreg_lag = np.divide(np.subtract(vs,vr_lag), vr_lag)
losses_lag = np.subtract(ps_lag,pr_lag)

E_lead = np.divide(pr_lead, ps_lead)
Vreg_lead = np.divide(np.subtract(vs,vr_lead), vr_lead)
losses_lead = np.subtract(ps_lead,pr_lead)
```
